```python
import sys, os

from codes.bsr_class_simulated_example import symreg
from codes.bsr_class_simulated_example import BSR
import pandas as pd
import numpy as np
import random
import time
import pickle
```

```python
def data(n=1000):
    sigma = 1.50 ** 0.50
    
    x1 = np.random.normal(4, 1, n)
    x2 = np.random.uniform(6, 1, n)
    x3 = np.random.uniform(8, 1, n)
    x1 = pd.DataFrame(x1)
    x2 = pd.DataFrame(x2)
    x3 = pd.DataFrame(x3)
    X = pd.concat([x1, x2, x3], axis=1)
    y = 5.00 * (X.iloc[:,0] + X.iloc[:,1]) * X.iloc[:,2] + np.random.normal(0, sigma, n)

    return X, y
```

```python
K = 2 # number of trees
MM = 100 # number of iterations
hyper_params = [{'treeNum': 2, 'itrNum':100, 'alpha1':0.4, 'alpha2':0.4, 'beta':-1.2}]
my_bsr = BSR(K,MM)
```

```python
my_bsr.fit(X, y)
model_expression = my_bsr.model()
complexity_BSR = my_bsr.complexity()
```

```python
data_rep = 1
rmse_BSR = [0] * (data_rep)
complexity_BSR_avg = [0] * (data_rep)
for d in range(data_rep):
    random.seed(d)

    n = 1000
    X, y = data(n)

    numiter = 25
    rmse = [0] * (numiter)
    complexity_BSR = [0] * (numiter)
    model_expression = [0] * (numiter)
    beta_hat = [0] * (numiter)
    y1 = y.to_numpy()
    #st = time.time()

    for i in range(numiter):
        my_bsr.fit(X, y)
        model_expression[i] = my_bsr.model()
        complexity_BSR[i] = my_bsr.complexity()
        fitted_y, beta_hat[i] = my_bsr.predict(X)
        fitted_y = fitted_y.flatten()
        rmse[i] = (np.sum((y1 - fitted_y) ** 2) ** 0.50)/(len(y1) ** 0.50)
        print(i)
    
    rmse_BSR[d] = min(rmse)
    complexity_BSR_avg[d] = np.mean(complexity_BSR)
    print(d)
```

```python
import sys, os

from codes.bsr_class_simulated_example import symreg
from codes.bsr_class_simulated_example import BSR
import pandas as pd
import numpy as np
import random
import time
import pickle
```

```python
def data(n=1000):
    sigma = 1.50 ** 0.50
    
    x1 = np.random.normal(4, 1, n)
    x2 = np.random.uniform(6, 1, n)
    x3 = np.random.uniform(8, 1, n)
    x1 = pd.DataFrame(x1)
    x2 = pd.DataFrame(x2)
    x3 = pd.DataFrame(x3)
    X = pd.concat([x1, x2, x3], axis=1)
    y = 5.00 * (X.iloc[:,0] + X.iloc[:,1]) * X.iloc[:,2] + np.random.normal(0, sigma, n)

    return X, y
```

```python
K = 2 # number of trees
MM = 100 # number of iterations
hyper_params = [{'treeNum': 2, 'itrNum':1000, 'alpha1':0.4, 'alpha2':0.4, 'beta':-1.2}]
my_bsr = BSR(K,MM)

data_rep = 25
runtime = [0.0] * data_rep
```

```python
for d in range(data_rep):
    random.seed(d)

    n = 1000
    X, y = data(n)

    X = np.asarray(X, dtype=float)
    y = np.asarray(y, dtype=float).reshape(-1)   # force 1-D

    numiter = 1
    #rmse = [0] * (numiter)
    #model_expression = [0] * (numiter)
    #beta_hat = [0] * (numiter)
    #y1 = y.to_numpy()
    #st = time.time()
    t0 = time.perf_counter()
    for i in range(numiter):
        my_bsr.fit(X, y)
        #model_expression[i] = my_bsr.model()
        #complexity_BSR[i] = my_bsr.complexity()
        #fitted_y, beta_hat[i] = my_bsr.predict(X)
        #fitted_y = fitted_y.flatten()
        #rmse[i] = (np.sum((y1 - fitted_y) ** 2) ** 0.50)/(len(y1) ** 0.50)
    
    t1 = time.perf_counter()
    runtime[d] = t1 - t0
    print(runtime[d])
    print('For d=:',d)
    #print(model_expression[rmse.index(min(rmse))])
```
