```python
y = pd.read_csv("y_catalysis")
X = pd.read_csv("X_catalysis")
X = X.drop(X.columns[0], axis = 1)
y = y.drop(y.columns[0], axis = 1)
y = y.squeeze(axis=0)
y = y.iloc[:,0]
```

```python
X_stage1 = X.loc[:, ['Hfo', 'Oxv']]
y_stage1 = y
```

```python
K = 2 # number of trees
MM = 100 # number of iterations
hyper_params = [{'treeNum': 2, 'itrNum':100, 'alpha1':0.4, 'alpha2':0.4, 'beta':-1.2}]
my_bsr = BSR(K,MM)
```

```python
numiter = 10
rmse = [0] * (numiter)
model_expression = [0] * (numiter)
beta_hat = [0] * (numiter)
y1 = y_stage1.to_numpy()
st = time.time()

for i in range(numiter):
    my_bsr.fit(X_stage1, y_stage1)
    model_expression[i] = my_bsr.model()
    fitted_y, beta_hat[i] = my_bsr.predict(X_stage1)
    fitted_y = fitted_y.flatten()
    rmse[i] = (np.sum((y1 - fitted_y) ** 2) ** 0.50)/(len(y) ** 0.50)
    print(i)
```

```python
data_rep = 25
K = [2, 3, 4, 5]
rmse_BSR = np.zeros((data_rep, len(K)))

for d in range(data_rep):
    
    X_train, X_test, y_train, y_test = train_test_split(X_stage1, y_stage1, test_size=0.1, random_state=d+1)
    X_train = X_train.reset_index(drop=True)
    X_test = X_test.reset_index(drop=True)
    y_train = y_train.reset_index(drop=True)
    y_test = y_test.reset_index(drop=True)
    
    for k in K:
        MM = 100 # number of iterations
        hyper_params = [{'treeNum': k, 'itrNum':100, 'alpha1':0.4, 'alpha2':0.4, 'beta':-1.2}]
        my_bsr = BSR(k,MM)
        numiter = 10
        rmse = [0] * (numiter)
        #model_expression = [0] * (numiter)
        beta_hat = [0] * (numiter)
        y1 = y_test.to_numpy()
    
        for i in range(numiter):
            my_bsr.fit(X_train, y_train)
            #model_expression[i] = my_bsr.model()
            fitted_y, beta_hat[i] = my_bsr.predict(X_test)
            fitted_y = fitted_y.flatten()
            rmse[i] = (np.sum((y1 - fitted_y) ** 2) ** 0.50)/(len(y1) ** 0.50)
            print("Iteration:", i)
    
        rmse_BSR[d][k-2] = min(rmse)
        print("Tree:", k)
    print("Data replication:", d+1)
```
