```python
import pandas as pd
import numpy as np
import random
import time
import feyn
```

```python
y = pd.read_csv("y_catalysis")
X = pd.read_csv("X_catalysis")
X = X.drop(X.columns[0], axis = 1)
y = y.drop(y.columns[0], axis = 1)
y = y.squeeze(axis=0)
y = y.iloc[:,0]
```

```python
X_stage1 = X.loc[:, ['Hfo', 'Oxv']]
y_stage1 = y
```

```python
dataset_stage1 = pd.DataFrame({
    'x1': X_stage1['Hfo'],
    'x2': X_stage1['Oxv'],
    'y': y_stage1
})
```

```python
# Train/test split
train, test = feyn.tools.split(dataset_stage1, ratio=[0.8, 0.2], random_state=42)

# Instantiate a QLattice
ql = feyn.QLattice(random_seed=69)

# Sample and fit models
models = ql.auto_run(
    data=train,
    output_name='y'
)

# Best model
best = models[0]

best.plot(train, test)
```

```python
data_rep = 25
rmse_QLattice = [0] * (data_rep)

X_stage1 = X.loc[:, ['Hfo', 'Oxv']]
y_stage1 = y

for d in range(data_rep):
    random.seed(d)

    # Create a DataFrame
    dataset_stage1 = pd.DataFrame({
    'x1': X_stage1['Hfo'],
    'x2': X_stage1['Oxv'],
    'y': y_stage1
    })
    
    # Train/test split
    train, test = feyn.tools.split(dataset_stage1, ratio=[0.9, 0.1], random_state=42)

    # Instantiate a QLattice
    ql = feyn.QLattice(random_seed=69)

    # Sample and fit models
    models = ql.auto_run(
        data=train,
        output_name='y'
    )

    # Best model
    best = models[0]

    #best.plot(train, test)
    rmse_QLattice[d] = (np.sum((test['y'] - best.predict(test)) ** 2) ** 0.50)/(len(test['y']) ** 0.50)

rmse_QLattice
```

```python
import csv

with open('test_rmse_QLattice_Single_Atom.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    for item in rmse_QLattice:
        writer.writerow([item])  # Write each item in a new row
```
